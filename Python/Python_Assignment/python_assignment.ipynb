```
def time_conversion(seconds): 
    seconds=seconds%(24*3600) 
    hours=seconds//3600
    seconds%=3600
    minutes=seconds//60
    seconds%=60
      
    return "%d:%02d:%02d" % (hours, minutes, seconds) 
  
time_in_seconds =1000
print(time_conversion(time_in_seconds))
```
